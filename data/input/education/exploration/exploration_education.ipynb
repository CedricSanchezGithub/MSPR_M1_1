# Exploration — Éducation

Exploration des fichiers éducation :
- **Diplômes** : `base-cc-diplomes-formation-2022.CSV`
- **CSP × Diplômes** : `pop-act2554-csp-dipl-cd-6822.xlsx`

```python
import pandas as pd
import os

DATA_DIR = os.path.join(os.path.dirname(os.getcwd()), '')  # dossier parent = education/
print(f"Dossier données : {DATA_DIR}")
```

## 1. Diplômes (RP 2022)

```python
dipl = pd.read_csv(os.path.join(DATA_DIR, 'base-cc-diplomes-formation-2022.CSV'), sep=';', low_memory=False)
print(f"Shape : {dipl.shape}")
print(f"Colonnes ({len(dipl.columns)}) :")
for c in dipl.columns[:20]:
    print(f"  - {c}")
if len(dipl.columns) > 20:
    print(f"  ... et {len(dipl.columns) - 20} autres")
dipl.head()
```

```python
dipl.info()
print()
dipl.describe()
```

```python
# Métadonnées
meta = pd.read_csv(os.path.join(DATA_DIR, 'meta_base-cc-diplomes-formation-2022.CSV'), sep=';', low_memory=False)
print(f"Shape métadonnées : {meta.shape}")
meta.head(20)
```

```python
# Fichier communes (COM)
com = pd.read_csv(os.path.join(DATA_DIR, 'base-cc-diplomes-formation-2022-COM.CSV'), sep=';', low_memory=False)
print(f"Shape COM : {com.shape}")
com.head()
```

## 2. CSP × Diplômes

```python
csp_dipl = pd.read_excel(os.path.join(DATA_DIR, 'pop-act2554-csp-dipl-cd-6822.xlsx'))
print(f"Shape : {csp_dipl.shape}")
print(f"Colonnes ({len(csp_dipl.columns)}) :")
for c in csp_dipl.columns[:20]:
    print(f"  - {c}")
if len(csp_dipl.columns) > 20:
    print(f"  ... et {len(csp_dipl.columns) - 20} autres")
csp_dipl.head()
```

## 3. Qualité des données

```python
for nom, df in [('Diplômes', dipl), ('CSP×Diplômes', csp_dipl)]:
    nulls = df.isnull().sum()
    total = nulls.sum()
    print(f"{nom} — {total} valeurs manquantes sur {df.shape[0] * df.shape[1]}")
    if total > 0:
        for col, n in nulls[nulls > 0].head(10).items():
            print(f"  {col}: {n} ({100*n/len(df):.1f}%)")
```
