# Exploration — Économie

Exploration des fichiers économiques :
- **Revenus** : `revenu-des-francais-a-la-commune-*.csv`
- **CSP** : `pop-act2554-csp-cd-6822.xlsx`
- **Secteurs d'activité** : `pop-act2554-empl-sa-sexe-cd-6822.xlsx`
- **Comptes communes** : `comptes_communes_*.csv` (4 fichiers)

```python
import pandas as pd
import os

DATA_DIR = os.path.join(os.path.dirname(os.getcwd()), '')  # dossier parent = economie/
print(f"Dossier données : {DATA_DIR}")
```

## 1. Revenus par commune

```python
rev_file = [f for f in os.listdir(DATA_DIR) if 'revenu' in f and f.endswith('.csv')][0]
revenus = pd.read_csv(os.path.join(DATA_DIR, rev_file), low_memory=False)
print(f"Fichier : {rev_file}")
print(f"Shape : {revenus.shape}")
print(f"Colonnes : {list(revenus.columns)}")
revenus.head()
```

```python
revenus.info()
print()
revenus.describe()
```

## 2. CSP (Catégories socio-professionnelles)

```python
csp = pd.read_excel(os.path.join(DATA_DIR, 'pop-act2554-csp-cd-6822.xlsx'))
print(f"Shape : {csp.shape}")
print(f"Colonnes ({len(csp.columns)}) :")
for c in csp.columns[:20]:
    print(f"  - {c}")
if len(csp.columns) > 20:
    print(f"  ... et {len(csp.columns) - 20} autres")
csp.head()
```

## 3. Secteurs d'activité

```python
secteurs = pd.read_excel(os.path.join(DATA_DIR, 'pop-act2554-empl-sa-sexe-cd-6822.xlsx'))
print(f"Shape : {secteurs.shape}")
print(f"Colonnes ({len(secteurs.columns)}) :")
for c in secteurs.columns[:20]:
    print(f"  - {c}")
if len(secteurs.columns) > 20:
    print(f"  ... et {len(secteurs.columns) - 20} autres")
secteurs.head()
```

## 4. Comptes des communes

```python
import glob

comptes_files = sorted(glob.glob(os.path.join(DATA_DIR, 'comptes_communes_*.csv')))
print(f"{len(comptes_files)} fichiers trouvés :")
for f in comptes_files:
    print(f"  {os.path.basename(f)}")
```

```python
# Aperçu du premier fichier
if comptes_files:
    cc = pd.read_csv(comptes_files[0], sep=';', low_memory=False)
    print(f"Fichier : {os.path.basename(comptes_files[0])}")
    print(f"Shape : {cc.shape}")
    print(f"Colonnes : {list(cc.columns)}")
    cc.head()
```

## 5. Qualité des données

```python
for nom, df in [('Revenus', revenus), ('CSP', csp), ('Secteurs', secteurs)]:
    nulls = df.isnull().sum()
    total = nulls.sum()
    print(f"{nom} — {total} valeurs manquantes sur {df.shape[0] * df.shape[1]}")
    if total > 0:
        for col, n in nulls[nulls > 0].head(10).items():
            print(f"  {col}: {n} ({100*n/len(df):.1f}%)")
```
